```python
import itertools
from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt

# Split genres
df_genres = df.assign(listed_in=df['Type'].str.split(',')).explode('Type')
df_genres['Type'] = df_genres['Type'].str.strip()

# Build pairs of co-occurring genres
pairs = []
for genres in df['Type'].dropna():
    genres = [g.strip() for g in genres.split(',')]
    pairs.extend(list(itertools.combinations(sorted(set(genres)), 2)))

pair_counts = Counter(pairs)
pair_df = pd.DataFrame(pair_counts.items(), columns=['Pair', 'Count'])
pair_df = pair_df[pair_df['Count'] > 100]  # show strong relationships only

# Build a network graph
G = nx.Graph()
for (g1, g2), c in pair_counts.items():
    if c > 100:
        G.add_edge(g1, g2, weight=c)

plt.figure(figsize=(10, 10))
nx.draw_networkx(G, with_labels=True, node_color='lightcoral', edge_color='gray')
plt.title("Genre Co-occurrence Network (Strong Genre Relationships)")
plt.show()
```

```python
df_cast_director = df.dropna(subset=['Director', 'Cast'])

pairs = []
for _, row in df_cast_director.iterrows():
    director = row['Director']
    for actor in str(row['Cast']).split(','):
        pairs.append((director.strip(), actor.strip()))

collab_df = pd.DataFrame(pairs, columns=['Director', 'Actor'])
top_collabs = collab_df.value_counts().head(15).reset_index(name='Collaboration_Count')

plt.figure(figsize=(10, 5))
sns.barplot(y=top_collabs['Director'] + " & " + top_collabs['Actor'],
            x=top_collabs['Collaboration_Count'], palette='rocket')
plt.title("Top 15 Director–Actor Collaborations on Netflix")
plt.xlabel("Collaboration Count")
plt.ylabel("Director–Actor Pair")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

yearly_titles = df.groupby('year_added').size().reset_index(name='count').dropna()
X = yearly_titles[['year_added']]
y = yearly_titles['count']

model = LinearRegression().fit(X, y)
next_year = pd.DataFrame({'year_added': [yearly_titles['year_added'].max() + 1]})
prediction = model.predict(next_year)[0]

print(f"Predicted number of new titles for {int(next_year.year_added)}: {int(prediction)}")
```
