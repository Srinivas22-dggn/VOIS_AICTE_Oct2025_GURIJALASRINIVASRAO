# EXPLORATORY DATA ANALYSIS

```python
df.describe()
```

```python
property_type=df["room type"].value_counts().to_frame()
```

```python
property_type
```

```python
property_type.plot(kind="bar")
plt.xticks(rotation=45)
for i in range(len(property_type)):
    plt.text(i, property_type.values[i] + 1, str(property_type.values[i]), ha='center', va='bottom')

plt.title("Property types and their count in the data")
plt.ylabel("room type count")
```

# neighbour hood group with highest no of listings

```python
hood_group=df["neighbourhood group"].value_counts()
```

```python
hood_group
```

```python
#plt.figure(figsize=(12,12))
plt.bar(hood_group.index,hood_group.values)


for i in range(len(hood_group)):
    plt.text(i, hood_group.values[i] + 1, str(hood_group.values[i]), ha='center', va='bottom')
plt.ylabel("no of listings")
plt.xlabel("Neighbourhood groups")
plt.title("Which Neighbourhood group has highest no of listings")

plt.show()
```

# which neighbourhood groups have highest average prices for airnb listings

```python
avg_price=df.groupby("neighbourhood group")["price"].mean().sort_values(ascending=False)
```

```python
plt.bar(avg_price.index,avg_price.values)
for i in range(len(hood_group)):
    plt.text(i, avg_price.values[i] + 1, str( int(avg_price.values[i])), ha='center', va='bottom')
plt.ylabel("Average price per listings $")
plt.xlabel("Neighbourhood groups")
plt.title("Which Neighbourhood group has highest average prices for airnb listings")

plt.show()
```

# is there a relationship between construction year and property of price

```python
df.groupby("Construction year")["price"].mean().plot(label="price")
plt.xlabel("construction year")
plt.ylabel("avearge price in $")
plt.legend()
plt.xticks(rotation=40)
plt.show()
```

# top 10 hosts by host listing count

```python
hosts=df.groupby("host name")["calculated host listings count"].sum().sort_values(ascending=False).nlargest(10)
```

```python
hosts
```

```python
plt.figure(figsize=(8,10))
plt.bar(hosts.index,hosts.values)
for i in range(len(hosts)):
    plt.text(i, hosts.values[i] + 1, str( int(hosts.values[i])), ha='center', va='bottom')
plt.ylabel("Calculated Host listings count")
plt.xlabel("Hosts name")
plt.title("Top 10 hosts calculated by host listing counts")
plt.xticks(rotation=45)
plt.show()
```

```python
review=df.groupby("host_identity_verified")["review rate number"].mean().sort_values(ascending=False)
```

```python
review
```

```python
review.plot(kind="bar")
for i in range(len(review)):
    plt.text(i, review.values[i] , str( review.values[i]), ha='center', va='bottom')
plt.ylabel("Avg review rate number")
plt.xlabel("host verfication status")
plt.title("Average review ratings for each verification statistics")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df,x="host_identity_verified",y="review rate number")
plt.xlabel("host verification status")
plt.ylabel("review rate number")
plt.title("average review rate for each verification status")
```
